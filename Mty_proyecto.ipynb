##Supervised: Clustering for text

```python
data= df_compras_enriched.copy()
```

> Reemplaza los valores faltantes (NaN) con la moda para variables categﾃｳricas y la mediana para variables numﾃｩricas, y convierte las variables categﾃｳricas en cﾃｳdigos numﾃｩricos. Veamos lo que cada lﾃｭnea de cﾃｳdigo hace:


Esta lﾃｭnea utiliza la funciﾃｳn enumerate para iterar sobre una lista de tipos de datos de cada columna del dataframe data.

Primero, se convierten los tipos de datos en una lista de strings usando la comprensiﾃｳn de lista [str(d) for d in data.dtypes], donde data.dtypes es una serie que contiene el tipo de datos de cada columna en el dataframe data. Esto se hace para asegurarse de que todos los elementos en la lista sean strings, lo cual es importante para que la siguiente lﾃｭnea de cﾃｳdigo funcione correctamente.

Luego, la funciﾃｳn enumerate itera sobre esta lista de tipos de datos y devuelve una tupla que contiene el ﾃｭndice del elemento y el valor del elemento en cada iteraciﾃｳn. Estas tuplas se desempaquetan en dos variables i y columns respectivamente.

De esta manera, en cada iteraciﾃｳn del ciclo for, i contiene el ﾃｭndice de la columna actual y columns contiene el tipo de dato de la columna actual en forma de string

```python
for i, columns in enumerate(list([str(d) for d in data.dtypes])):
#Este bucle for que recorre todas las columnas en el conjunto de datos data. i es el ﾃｭndice de la columna y columns es el nombre de la columna.
  
  if columns == 'object':
    data[data.columns[i]]= data[data.columns[i]].fillna(data[data.columns[i]].mode())
    data[data.columns[i]]= data[data.columns[i]].astype('category').cat.codes

#Esta condiciﾃｳn if comprueba si el tipo de la columna es una variable categﾃｳrica. 
#Si la columna es categﾃｳrica,entonces los valores faltantes (NaN) se reemplazan con la moda utilizando fillna(data[data.columns[i]].mode()). 
#Luego, los valores categﾃｳricos se convierten en cﾃｳdigos numﾃｩricos utilizando astype('category').cat.codes    
  
  else:
    data[data.columns[i]]= data[data.columns[i]].fillna(data[data.columns[i]].median())
#Si la columna no es categﾃｳrica, entonces los valores faltantes se reemplazan con la mediana utilizando fillna(data[data.columns[i]].median()). 
#La mediana se usa porque es una medida resistente a los valores atﾃｭpicos y puede ser una mejor opciﾃｳn que la media en algunos casos.
```

##Unsupervised: Anomalias referente a importes

```python
import gensim
from nltk.corpus import stopwords
#np.random.seed(2)
```

```python
#tokens
def preprocess(text):
  result= []
  for token in text.split():
    if token not in stopwords.words('spanish') and len(token)>4:
      result.append(token)
  return result
```

```python
#cluster busines
process_business = [preprocess(x) for x in list(df_compras_enriched['BENEFICIARIO'].unique())]
```

```python
dictionary= gensim.corpora.Dictionary(process_business)
```

```python
#filtro quitando ocurrencias frecuentes y poco frecuentes
dictionary.filter_extremes(no_below=1, no_above=0.5)
```

```python
bow_corpus= [dictionary.doc2bow(doc) for doc in process_business]
```

```python
lda_model= gensim.models.LdaMulticore(bow_corpus, num_topics=6, id2word=dictionary, passes=10, workers=2)
```

```python
for idx, type_business in lda_model.print_topics(-1):
  print('Topic: {} words {}'.format(idx, type_business))
```

```python
def get_max_topic(index):
  results= lda_model[bow_corpus[index]]
  return max(lda_model[results], key=lambda x: x[1])[0]
```

```python
business_cluster = {x: get_max_topic(i) for i, x in enumerate(list(df_compras_enriched['BENEFICIARIO'].unique() ))}
```

```python
df_compras_enriched['CLUSTER_NAME']= df_compras_enriched['BENEFICIARIO'].apply(lambda x:business_cluster[x])
```

```python
df_compras_enriched.groupby('CLUSTER_NAME')['IMPORTE'].mean().plot.bar()
```

##Aplicaciﾃｳn unsupervised: Anomalias

```python
from scipy.stats import norm
import numpy as np
from scipy.stats import norm

# Eliminar o imputar valores nulos o NaN
df_compras_enriched = df_compras_enriched.dropna(subset=['IMPORTE'])

# Convertir la columna 'IMPORTE' a un tipo de datos numﾃｩrico
df_compras_enriched['IMPORTE'] = df_compras_enriched['IMPORTE'].astype(float)

# Ajustar el modelo de distribuciﾃｳn normal
params = norm.fit(df_compras_enriched['IMPORTE'])

# Calcular la probabilidad acumulada
df_compras_enriched['PROB_TOTAL'] = norm.cdf(df_compras_enriched['IMPORTE'], params[0], params[1])
```

```python
df_compras_enriched.PROB_TOTAL.hist(bins=5)
```

```python
df_compras_enriched[df_compras_enriched['PROB_TOTAL']>.95]
```

```python
models_sheet={}
for sheet in list(df_compras_enriched['SHEET'].unique()):
  sheet_compras=df_compras_enriched[df_compras_enriched['SHEET']== sheet]
  params= norm.fit(sheet_compras['IMPORTE'])
  models_sheet[sheet]= norm(params[0], params[1])
```

```python
sheet_prob = []
for i, row in df_compras_enriched.iterrows():
  sheet_prob.append(models_sheet[row['SHEET']].cdf(row['IMPORTE']))
df_compras_enriched['PROB_SHEET']= sheet_prob
```

```python
df_compras_enriched['PROB_SHEET'].hist(bins=5)
```

```python
df_compras_enriched[df_compras_enriched['PROB_SHEET']>0.95]
```
